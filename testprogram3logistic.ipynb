```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
pregnancy_data = pd.read_csv('/content/bootcamp.csv')
pregnancy_data.head()
pregnancy_data.tail()
pregnancy_data.shape
pregnancy_data.info()
pregnancy_data.isnull().sum()
pregnancy_data.describe()
pregnancy_data['RiskLevel'].value_counts()
sns.set()
plt.figure(figsize=(6,6))
sns.distplot(pregnancy_data['Age'])
plt.title('Age of pregnant women')
plt.show()
plt.figure(figsize=(6,6))
sns.distplot(pregnancy_data['SystolicBP'])
plt.title('Systolic BP')
plt.show()
plt.figure(figsize=(6,6))
sns.distplot(pregnancy_data['DiastolicBP'])
plt.title('Diastolic BP')
plt.show()
plt.figure(figsize=(6,6))
sns.distplot(pregnancy_data['BS'])
plt.title('BS distribution')
plt.show()
plt.figure(figsize=(6,6))
sns.distplot(pregnancy_data['BodyTemp'])
plt.title('Temperature')
plt.show()
plt.figure(figsize=(6,6))
sns.distplot(pregnancy_data['HeartRate'])
plt.title('Heart Rate')
plt.show()
plt.figure(figsize=(6,6))
sns.countplot(x='RiskLevel', data=pregnancy_data)
plt.title('RiskLevel')
plt.show()
X = pregnancy_data.drop(columns='RiskLevel', axis=1)
Y = pregnancy_data['RiskLevel']
print(X)
print(Y)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
model = LogisticRegression()
model.fit(X_train, Y_train)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
input_data = (19,135,71,7.01,78,68)


input_data_as_numpy_array= np.asarray(input_data)


input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The pregnant patient is at low risk')
else:
  print('The pregnant patient is at high risk')
```
